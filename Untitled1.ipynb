```python
import cv2 as cv
import sys
import requests
import numpy as np
from io import BytesIO
from google.colab.patches import cv2_imshow

url = "https://drive.google.com/uc?export=download&id=1pmut2I2iyVcMX4jfHKmUPOBr8pcQIr28"
response = requests.get(url)
if response.status_code != 200:
    sys.exit("Failed to retrieve the image from the URL.")
image_bytes = BytesIO(response.content)
img = cv.imdecode(np.frombuffer(image_bytes.read(), np.uint8), cv.IMREAD_COLOR)
if img is None:
    sys.exit("Could not read the image.")
max_width = 500
max_height = 300
height, width = img.shape[:2]
scale_factor = min(max_width / width, max_height / height)
new_width = int(width * scale_factor)
new_height = int(height * scale_factor)
resized_img = cv.resize(img, (new_width, new_height))
cv2_imshow(resized_img)
```
