```python
import cv2 as cv
import sys
import requests
import numpy as np
from io import BytesIO
from google.colab.patches import cv2_imshow  # For Colab image display

# Google Drive direct download link (replace file ID with yours)
url = "https://drive.google.com/uc?export=download&id=1OvG6up584bTLWS-sA1BGO1UpGfpMGp3_"

# Download the image
response = requests.get(url)

if response.status_code != 200:
    sys.exit("Failed to retrieve the image from the URL.")

# Convert downloaded bytes into an OpenCV image (grayscale)
image_bytes = BytesIO(response.content)
img = cv.imdecode(np.frombuffer(image_bytes.read(), np.uint8), cv.IMREAD_GRAYSCALE)

if img is None:
    sys.exit("Could not read the image.")

# Resize image to fit within max dimensions while keeping aspect ratio
max_width = 500
max_height = 300

height, width = img.shape[:2]
scale_factor = min(max_width / width, max_height / height)

new_width = int(width * scale_factor)
new_height = int(height * scale_factor)

resized_img = cv.resize(img, (new_width, new_height))

# Display the grayscale image in Colab
cv2_imshow(resized_img)
```
